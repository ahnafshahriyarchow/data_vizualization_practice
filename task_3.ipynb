**The objective of this question is to explore correlations, non-linear relationships, and patterns in the Titanic dataset by using visual techniques.**





```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic_data = pd.read_csv(url)
titanic_data.dropna(inplace=True)
```

 **(a)** Implement a function to generate a heatmap to examine the correlation between continuous/numeric vari-
ables. Use the ‘hue‘ argument to color-code data by ‘species‘.

```python
#Solution to 3(a)

def plot_correlation_heatmap(df_data):
    plt.figure(figsize=(10, 8))
    correlation_matrix = df_data.corr(numeric_only=True)
    sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
    plt.title("Titanic Data Correlation Heatmap")
    plt.show()
    
```

```python
plot_correlation_heatmap(titanic_data)
```

 **(b)** Implement a function that creates scatter plots matrix to explore potential non-linear relationships between variables.

```python
#Solution to 3(b)

def plot_scatter_matrix(df_data):
    hue_col = 'Pclass' if 'Pclass' in df_data.columns else None
    sns.set(style="darkgrid")
    g = sns.pairplot(df_data, hue=hue_col, diag_kind="kde", palette="magma") if hue_col else sns.pairplot(df_data, diag_kind="kde", palette="magma")
    plt.show()
```

```python
plot_scatter_matrix(titanic_data)
```

 **(c)** Implement a function to generate scatter plots comparing ‘age‘ and ‘fare‘ for each passenger class

```python
#Solution to 3(c)

def plot_age_vs_fare(df_data):
    plt.figure(figsize=(8, 6))
    pclass_labels = df_data["Pclass"].unique() 
    idx = 0
    while idx < len(pclass_labels):
        pclass = pclass_labels[idx]
        subset = df_data[df_data["Pclass"] == pclass]
        plt.scatter(subset["Age"], subset["Fare"], label=f"{pclass}", alpha=0.7)
        idx += 1

    plt.xlabel("Age")
    plt.ylabel("Fare")
    plt.legend(title="PClass")
    plt.title("Scatter Plot of Age vs Fare by Passenger Class")
    plt.show()
```

```python
plot_age_vs_fare(titanic_data)
```
