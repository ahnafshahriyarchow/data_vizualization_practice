**The objective of this question is to analyze the distribution and density of data using appropriate
visual techniques. You will be working with the same penguins dataset from task 1.
Task: Implement a function to generate histograms or density plots to examine the distribution of any provided continuous
variable from the dataset. The function should take inputs for the column to be plotted and the type of plot (histogram or
density plot).**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
penguin_data = pd.read_csv(url)
penguin_data.dropna(inplace=True)
```

```python
def plot_univariate_distribution(df_data, column_name, plot_type="histogram"):
    
    plt.figure(figsize=(8, 6))
    if plot_type == "histogram":
        plt.hist(df_data[column_name], bins=30, edgecolor="black", alpha=0.7)
    elif plot_type == "density":
        sns.kdeplot(df_data[column_name], fill=True, alpha=0.6)

    plt.xlabel(column_name)
    plt.ylabel("Frequency" if plot_type == "histogram" else "Density")
    plt.title(f"{plot_type.capitalize()} Plot of {column_name}")
    plt.show()
```

```python
plot_univariate_distribution(penguin_data, "bill_length_mm", plot_type="histogram")
```

```python
plot_univariate_distribution(penguin_data, "body_mass_g", plot_type="density")
```
