**In this question you are tasked with visualizing mnist fashion data in 2D.**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import umap
from tensorflow.keras.datasets import fashion_mnist
```

```python
def load_fashion_mnist(sample_size=5000):
    (train_images, train_labels), (_, _) = fashion_mnist.load_data()
    train_images_flat = train_images.reshape(-1, 28*28)
    pixel_values_sample = train_images_flat[:sample_size]
    targets_sample = train_labels[:sample_size]
    return pixel_values_sample, targets_sample
```

```python
def apply_umap(pixel_values_sample, n_components=2):
    reducer = umap.UMAP(n_components=n_components, random_state=42)
    transformed_data = reducer.fit_transform(pixel_values_sample)
    return transformed_data
```

```python
def plot_umap_embedding(transformed_data, targets):
    plt.figure(figsize=(8, 12))
    sns.scatterplot(x=transformed_data[:, 0], y=transformed_data[:, 1], hue=targets, palette="tab10", s=30, edgecolor=None)
    plt.legend(title="targets", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Fashion MNIST Data in 2D")
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.show()
```

```python
pixel_values_sample, targets_sample = load_fashion_mnist(sample_size=5000)
transformed_data = apply_umap(pixel_values_sample)
plot_umap_embedding(transformed_data, targets_sample)
```
