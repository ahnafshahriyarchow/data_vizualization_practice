** In this question you will work with the penguins dataset to examine the importance of visual variables in effective data visualization.**

 **(a)** Implement a basic scatter plot as specified in the below code samples. This scatter plot will show the relationship between any provided two features, with "species" differentiated by color.  

 **(b)** Implement an enhanced scatter plot as specified in the code sample below. This scatter plot will show the relationship between any provided two features, with the "species" differentiated by color, point size representing another feature, and an optional "style" (shape) that can represent additional grouping, e.g., "sex." The function should take inputs for X and Y axes, as well as the columns for color (hue), point size, and style.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
penguin_data = pd.read_csv(url)
penguin_data.dropna(inplace=True)
```

```python
#Solution to 1(a)

def create_scatter_plot_basic(df_data, x_axis_value, y_axis_value, hue_value="species", outfile_name="q1a_penguins_scatter_1.png"):
    
    plt.figure(figsize=(8, 6))
    unique_species = df_data[hue_value].unique() 
    idx = 0
    while idx < len(unique_species):
        species = unique_species[idx]
        species_data = df_data[df_data[hue_value] == species]
        plt.scatter(species_data[x_axis_value], species_data[y_axis_value], label=species, alpha=0.7)
        idx += 1
    plt.xlabel(x_axis_value)
    plt.ylabel(y_axis_value)
    plt.legend(title=hue_value)
    plt.title(f"Scatter Plot: {x_axis_value} vs {y_axis_value}")
    plt.show()
```

```python
create_scatter_plot_basic(penguin_data, "bill_length_mm", "bill_depth_mm")
```

```python
#Solution to 1(b)

def create_scatter_plot_advanced(df_data, x_axis_value, y_axis_value, size_value, style_value, hue_value="species", outfile_name="q1b_penguins_scatter2.png"):
    
    plt.figure(figsize=(10, 8))
    sns.scatterplot(data=df_data, x=x_axis_value, y=y_axis_value, hue=hue_value, size=size_value, style=style_value, palette="tab10", sizes=(30, 150), edgecolor='white')
    plt.title("Enhanced Scatter Plot: flipper_length_mm vs body_mass_g with bill_length_mm by species and sex")
    plt.xlabel("body_mass_g")
    plt.ylabel("flipper_length_mm") 
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.show()
```

```python
create_scatter_plot_advanced(penguin_data, "flipper_length_mm", "body_mass_g", "bill_length_mm", "sex")
```
